# 1. PREPARACIÓN DEL ENTORNO

Comenzaremos descargando una imagen de ejemplo para nuestro análisis. Trabajaremos con una imagen de frutos rojos que nos permitirá apreciar claramente las diferencias entre canales de color.

```python
import urllib.request

# Descarga de imagen de ejemplo
url = "https://www.dropbox.com/s/do6b49d10ty1dyf/frutos_rojos.png?dl=1"
ruta_destino = "frutos_rojos.png"  # Se guarda en la carpeta actual

urllib.request.urlretrieve(url, ruta_destino)

print("📁 Imagen descargada correctamente")
print("🎯 Vamos a trabajar con una imagen de frutos rojos para entender los canales de color")
```

# 2. OPENCV: LA HERRAMIENTA ESTÁNDAR DE VISIÓN COMPUTACIONAL

**NOTA TÉCNICA IMPORTANTE:**
OpenCV usa el formato BGR (Azul-Verde-Rojo) en lugar de RGB. Esto es por razones históricas del desarrollo de cámaras y software.

A continuación, cargaremos la imagen con OpenCV y analizaremos sus dimensiones antes de visualizarla.

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
# Carga de imagen con OpenCV
X = cv2.imread('frutos_rojos.png')
print(f"Dimensiones de la imagen: {X.shape}")
print("   - Alto x Ancho x Canales")
print("   - El tercer valor (3) indica que es una imagen en color")

# Mostrar imagen en Jupyter/VS Code
plt.imshow(cv2.cvtColor(X, cv2.COLOR_BGR2RGB))
plt.title("Imagen con OpenCV (convertida a RGB)")
plt.axis('off')
plt.show()
```

# 3. ANÁLISIS DE CANALES DE COLOR (OpenCV - Formato BGR)

**Recordemos**: en OpenCV el orden es B-G-R (no R-G-B como estamos acostumbrados)

Ahora separaremos cada canal para visualizar su contribución independiente a la imagen.
- **Canal Azul (B)**: En los frutos rojos, este canal será oscuro. Los píxeles rojos tienen poco componente azul.
- **Canal Verde (G)**: Componente intermedio en frutos rojos. Las hojas verdes se verán brillantes aquí.
- **Canal Rojo (R)**: Los frutos rojos se verán muy brillantes. Este es el canal dominante en nuestra imagen.

```python
print("🎨 ANÁLISIS POR CANALES - Formato BGR de OpenCV")

# Canal Azul (Blue) - Índice 0
print("\n🔵 CANAL AZUL (B)")
B = X[:,:,0]
plt.imshow(B, cmap='gray')
plt.title("Canal Azul (B)")
plt.axis('off')
plt.show()

# Canal Verde (Green) - Índice 1
print("\n🟢 CANAL VERDE (G)")
G = X[:,:,1]
plt.imshow(G, cmap='gray')
plt.title("Canal Verde (G)")
plt.axis('off')
plt.show()

# Canal Rojo (Red) - Índice 2
print("\n🔴 CANAL ROJO (R)")
R = X[:,:,2]
plt.imshow(R, cmap='gray')
plt.title("Canal Rojo (R)")
plt.axis('off')
plt.show()
```

# 4. MATPLOTLIB: HERRAMIENTA DE VISUALIZACIÓN CIENTÍFICA

Matplotlib es una herramienta de visualización científica que maneja imágenes en formato RGB (más intuitivo).

A diferencia de OpenCV, Matplotlib lee automáticamente las imágenes en formato RGB, lo que facilita la visualización directa sin conversiones adicionales.

```python
# Carga con matplotlib
X_plt = plt.imread('frutos_rojos.png')
print(f"Dimensiones: {X_plt.shape}")
print("Nota: Matplotlib lee automáticamente en formato RGB")

plt.figure(figsize=(10, 8))
plt.imshow(X_plt)
plt.title('Imagen Original - Formato RGB (Matplotlib)', fontsize=14, pad=20)
plt.axis('off')
plt.show()
```

Si el tercer valor es igual a 4, significa que la imagen tiene cuatro canales.
Normalmente, el cuarto canal es el canal alfa, que representa la transparencia de cada píxel (además de los tres canales de color: azul, verde y rojo).

Por ejemplo, una imagen con dimensiones (512, 512, 4) tiene:

512 píxeles de alto
512 píxeles de ancho
4 canales: B (azul), G (verde), R (rojo) y A (alfa/transparencia)
Esto es común en imágenes PNG con transparencia.

# 5. ANÁLISIS DE CANALES CON MATPLOTLIB (Formato RGB)

Ahora analizaremos los canales RGB utilizando Matplotlib. Observen cómo difiere el orden de los canales respecto a OpenCV:
- En RGB: el primer canal (índice 0) es el Rojo
- En RGB: el segundo canal (índice 1) es el Verde
- En RGB: el tercer canal (índice 2) es el Azul

```python
print("ANÁLISIS POR CANALES - Formato RGB de Matplotlib")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Canal Rojo (Red) - Índice 0 en RGB
R_plt = X_plt[:,:,0]
axes[0].imshow(R_plt, cmap='gray')
axes[0].set_title('Canal ROJO\n(Los frutos se ven brillantes)', fontsize=12)
axes[0].axis('off')

# Canal Verde (Green) - Índice 1 en RGB
G_plt = X_plt[:,:,1]
axes[1].imshow(G_plt, cmap='gray')
axes[1].set_title('Canal VERDE\n(Las hojas se ven brillantes)', fontsize=12)
axes[1].axis('off')

# Canal Azul (Blue) - Índice 2 en RGB
B_plt = X_plt[:,:,2]
axes[2].imshow(B_plt, cmap='gray')
axes[2].set_title('Canal AZUL\n(Los frutos se ven oscuros)', fontsize=12)
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

# 6. CONVERSIÓN ENTRE FORMATOS: BGR ↔ RGB

**Problema común**: OpenCV lee BGR pero matplotlib espera RGB.

**Solución**: Usar cv2.cvtColor() para convertir entre formatos.

A continuación demostraremos el problema que surge al visualizar directamente una imagen en formato BGR con Matplotlib (que espera RGB), y cómo solucionarlo mediante conversión.

```python
# Leemos con OpenCV (BGR) y convertimos a RGB
Xbgr = cv2.imread('frutos_rojos.png')
Xrgb = cv2.cvtColor(Xbgr, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Sin conversión (colores incorrectos)
axes[0].imshow(Xbgr)
axes[0].set_title('Sin Conversión\n(OpenCV BGR → Matplotlib)', fontsize=12, color='red')
axes[0].axis('off')

# Con conversión (colores correctos)
axes[1].imshow(Xrgb)
axes[1].set_title('Con Conversión\n(BGR → RGB)', fontsize=12, color='green')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
